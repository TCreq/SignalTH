```python
"""Exemple simple de manipulation d'images."""

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import math as m

image = Image.open("chardin-la-raie-1728.jpg")
image = image.convert('L')

print(image.entropy())
image
```

```python
M=np.asarray(image)
M.size
```

```python
ll=[[u for u in l] for l in M]
L=[]
for l in ll:
    L+=l
L=set(L)
L
D={}
for u in L:
    D[u]=sum(sum(M==u))
D
```

```python
mylist = [key for key, val in D.items() for _ in range(val)]
plt.hist(mylist, bins=238)
```

```python
ll=[[u for u in l] for l in M]
L=[]
for l in ll:
    L+=l
L=set(L)
L
D1={}
for u in L:
    D1[u]=sum(sum(M==u))/M.size
D1
```

```python
H=sum([-D1[j]*m.log2(D1[j]) for j in D1])
H
```
