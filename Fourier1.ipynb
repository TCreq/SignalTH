```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```python
A=1
T=1
tau=0.5
def u(t):
    return A*((t%T<(tau/2))|((t%T>(T-tau/2))))

tps=np.linspace(-1,3,1000)
U=u(tps)

plt.plot(tps,U)
```

```python
per=np.linspace(0,T,1000)
pas=per[1]-per[0]

def a(n):
    return sum(u(per)*np.cos(2*m.pi*n*per/T)*pas)*2/T*(1/2*(n==0)+1*(n!=0))
    
def b(n):
    return sum(u(per)*np.sin(2*m.pi*n*per/T)*pas)*2/T
    
    
```

```python
def v(t,n=1):
    def cst(t0):
        return a(0)
    vi=cst(t)
    for i in range(1,n+1):
        vi+=a(i)*np.cos(2*m.pi*i*t/T)+b(i)*np.sin(2*m.pi*i*t/T)
    return vi
        
```
